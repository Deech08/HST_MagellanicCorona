```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

import astropy.units as u

import seaborn as sns
pal = sns.color_palette("colorblind")
```

```python
def M200(T):
    return 1537.7*T**(3/2)
    
def Tvir(M200):
    return (M200 / 1537.7)**(2/3)
```

```python
T = np.logspace(4,6.5,1000)
```

```python
fig,ax = plt.subplots()

ax.plot(np.log10(T), np.log10(M200(T)), lw = 2, label = "Virial Relation", color = "k", alpha = 0.7)

ax.axhline(np.log10(14*1e10), ls = "--", label  = "Erkal+ 2019", lw = 2, alpha = 0.7, color = pal[3])
ax.axvline(np.log10(Tvir(14*1e10)), ls = "--", lw = 2, alpha = 0.7, color = pal[3])

ax.axhline(np.log10(0.25*1e12), ls = ":", label = "Penarrubia+ 2016", lw = 2, alpha = 0.7, color = pal[0])
ax.axvline(np.log10(Tvir(0.25*1e12)), ls = ":", lw = 2, alpha = 0.7, color = pal[0])

ax.axvline(4.96, ls = "-.", alpha = 0.7, label = "Measured T (primary)", lw = 2, color = pal[2])
ax.axvline(5.5, linestyle = "-.", alpha = 0.7, lw = 2, color = pal[4], 
           label = "Measured T (secondary)")

lg = ax.legend(fontsize = 10)

ax.set_xlabel(r"$\log_{10}(T/\mathrm{K})$", fontsize = 12)
ax.set_ylabel(r"$\log_{10}(M/M_\odot)$", fontsize = 12)

ax.set_xlim(4,6.5)
ax.set_ylim(9.18,12.95)

fig.savefig("/Users/dk/Dropbox/STScI/Research/MagCorona/ScienceFigures/VirialRelations.png", dpi = 300)
```
