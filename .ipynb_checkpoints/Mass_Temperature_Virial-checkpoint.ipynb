```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

import astropy.units as u
```

```python
def M200(T):
    return 1537.7*T**(3/2)
    
```

```python
T = np.logspace(4,6.5,1000)
```

```python
fig,ax = plt.subplots()

ax.plot(np.log10(T), np.log10(M200(T)))
```
